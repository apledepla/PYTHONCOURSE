```python
import requests


# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')
```

```python
import time

print('Number 2:')

def fibo_iter(n):
    x = 1
    z = 0
    date = 0
    for i in range(n):
        date = z
        z = x + z
        x = date
    return z

def fibo_recur(n, x=1, z=0):
    if n != 0:
        return fibo_recur(n-1, z, x+z)
    else:
        return z
    
n = int(input('Enter number of fibonacci sequence: '))

start_time = time.time()   
fibo_iter(n)
end_time = time.time()
print("Fibonacci sequence (Iteration): ", fibo_iter(n))
print("Time taken (Iteration): ", end_time-start_time, "seconds")



start_time = time.time()
print("\nFibonacci sequence (Recursion): ", fibo_recur(n))
end_time = time.time()
print("Time taken (Recursion): ", end_time-start_time, "seconds")
```
